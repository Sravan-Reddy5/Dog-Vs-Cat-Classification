```python
from tensorflow.keras.models import Sequential
```

MODEL

```python
cnn = tensorflow.keras.models.Sequential([
    Conv2D(filters=256, kernel_size=(3, 3), activation='relu', input_shape=(224,224, 3)),
    MaxPooling2D((2, 2)),
    
    Conv2D(filters=128, kernel_size=(2, 2), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(filters=64, kernel_size=(2, 2), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(filters=32, kernel_size=(2, 2), activation='relu'),
    Flatten(),
    Dense(256,activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(2, activation='softmax')
])
```

```python
import tensorflow as tf
```

```python
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = cnn.fit(X_train, Y_train, epochs=15,validation_split=0.1)
```

```python
del cnn
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel('epochs')
plt.legend(['training_data','valaidation data'],loc='lower right')
```
